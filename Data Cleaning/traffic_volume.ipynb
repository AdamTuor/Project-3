```python
import pandas as pd
import datetime as dt
```

```python
data09 = pd.read_csv('../Resources/raw-data-2000-2009.csv')
data19 = pd.read_csv('../Resources/raw-data-2010-2019.csv')
data29 = pd.read_csv('../Resources/raw-data-2020-2029.csv')
```

```python
df09 = pd.DataFrame(data09)
df19 = pd.DataFrame(data19)
df29 = pd.DataFrame(data29)
```

```python
#converting the date from string to datetime
df09['count_date']=pd.to_datetime(df09['count_date'])
```

```python
df09.dtypes
```

```python
#filter for the years we want
filter = (df09['count_date'].dt.year >= 2007)
filtered_df09 = df09.loc[filter]
```

```python
#adding totals to make groupby easier
filtered_df09['total_cars'] = filtered_df09.loc[:, 'sb_cars_r':'eb_bus_l'].sum(axis=1)
filtered_df09['total_pedestrians'] = filtered_df09.loc[:, 'nx_peds':'wx_peds'].sum(axis=1)
filtered_df09['total_bikes'] = filtered_df09.loc[:, 'nx_bike':'wx_bike'].sum(axis=1)
```

```python
#drop all of the individual columns for readability
filtered_df09 = filtered_df09.drop(columns=['sb_cars_r', 'sb_cars_t', 'sb_cars_l', 'nb_cars_r',
       'nb_cars_t', 'nb_cars_l', 'wb_cars_r', 'wb_cars_t', 'wb_cars_l',
       'eb_cars_r', 'eb_cars_t', 'eb_cars_l', 'sb_truck_r', 'sb_truck_t',
       'sb_truck_l', 'nb_truck_r', 'nb_truck_t', 'nb_truck_l', 'wb_truck_r',
       'wb_truck_t', 'wb_truck_l', 'eb_truck_r', 'eb_truck_t', 'eb_truck_l',
       'sb_bus_r', 'sb_bus_t', 'sb_bus_l', 'nb_bus_r', 'nb_bus_t', 'nb_bus_l',
       'wb_bus_r', 'wb_bus_t', 'wb_bus_l', 'eb_bus_r', 'eb_bus_t', 'eb_bus_l',
       'nx_peds', 'sx_peds', 'ex_peds', 'wx_peds', 'nx_bike', 'sx_bike',
       'ex_bike', 'wx_bike', 'nx_other', 'sx_other', 'ex_other', 'wx_other'])
```

```python
filtered_df09.head()
```

```python
#group by location, year, lat/lng and return just the counts we're interested in for our traffic volume
grouped09 = filtered_df09.groupby(['location',filtered_df09['count_date'].dt.year,'lng','lat'])[['total_cars','total_pedestrians','total_bikes']].sum().reset_index()
```
